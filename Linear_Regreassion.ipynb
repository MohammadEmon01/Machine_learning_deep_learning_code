# what is Linear Regression

 Linear Regression is a supervised machine learning algorithm used for regression tasks. It models the relationship between a dependent variable and one or more independent variables by fitting a linear equation to observed data.
 types of linear regression:
    # simple regression: only one independent variabble 
    # Muliple regression: more than independent variables 
    



```python
# import libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split    
from sklearn.metrics import mean_squared_error,root_mean_squared_error,mean_absolute_error
```

```python
#load the dataset 
df = sns.load_dataset('tips')
```

```python
df
```

```python
# drow scatterplot 
sns.scatterplot(x='total_bill',y='tip',data=df)
```

```python
#split the dataset into features and target variable
x = df[['total_bill']]
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
y = df['tip']
```

```python
# train test split 
x_train, x_test, y_train , y_test = train_test_split(x,y,test_size = 0.2)
```

```python
# call the model 
model = LinearRegression()
```

```python
# train the model 
model.fit(x_train,y_train)
```

```python
# take out the intercept and slope 
print (model.intercept_)
print(model.coef_)
#print ('y = mx + b')
#print ('y = {}x + {}'.format(model.coef_[0],model.intercept_))
print('y =',model.intercept_, '+',model.coef_,'*x')
```

```python
y_pred = model.predict(x_test)
```

```python
print('MSE=',mean_squared_error(y_test,y_pred))

print('RMSE=',np.sqrt(mean_squared_error(y_test,y_pred)))
```
