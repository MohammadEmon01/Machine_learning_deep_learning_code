```python
#Feature encaoding
#Freature encoding is a technique used to convert "categorical variables" into "numerical values" that can be used in machine learning algorithms. This is important because most machine learning algorithms work with numerical data and cannot directly handle categorical variables.
#Types of feature encoding
#1. Label Encoding
#2. One-Hot Encoding
#3. Binary Encoding
#4. frequency Encoding
```

```python
#import libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
df['time'].value_counts()
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,OrdinalEncoder
```

```python
Le = LabelEncoder()
df['Time_enconded']=Le.fit_transform(df['time'])
```

```python
df['Time_enconded'].value_counts()
```

```python
df.head(5)
```

```python
#OrdinalEncoder
Oe = OrdinalEncoder(categories=[['Thur','Fri','Sat','Sun']])
df['Day_encoded']=Oe.fit_transform(df[['day']])
```

```python
df['Day_encoded'].value_counts()
```

```python
#One-Hot Encoding
Ohe = OneHotEncoder()
Ohe.fit_transform(df[['smoker']]).toarray()
```

Discretization

## Data discretization is the process of converting continuous data into discrete categories or bins. This is often done to simplify the data and make it easier to analyze or visualize. Discretization can be useful in various scenarios, such as when you want to reduce noise in the data, improve model performance, or create interpretable features.

```python
import seaborn as sns 
import pandas as pd 
from sklearn.preprocessing import KBinsDiscretizer

```

```python
df = sns.load_dataset('titanic')
df.head(5)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['age']=df['age'].ffill()
```

```python
df.isnull().sum()
```

```python
sns.histplot(df['age'],kde=True)
```

```python
#age discretization
age_dis = KBinsDiscretizer(n_bins=2, encode='ordinal', strategy='uniform')
df['age_bin']= age_dis.fit_transform(df[['age']])
```

```python
df.head(5)
```

```python
sns.histplot(df,x='age',hue='age_bin')
```
