```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
#what is Standard scaling?
#Standard scaling is a method of normalizing data by removing the mean and scaling to unit variance. It transforms the data such that it has a mean of 0 and a standard deviation of 1. This is particularly useful when the features have different units or scales, as it ensures that each feature contributes equally to the analysis.
#Standard scaling is commonly used in machine learning algorithms that are sensitive to the scale of the data, such as k-means clustering and support vector machines.
#The Furmula for Standard Scaling is:
#X_scaled = (X - mean) / std
#where X is the original data, mean is the mean of the data, and std is the standard deviation of the data.
```

```python
#import libraries
import pandas as pad 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#Generate a random dataset(Exponential distribution)
#please explain what is seed and why we use it in this code
#seed is a number that initializes the random number generator. It ensures that the random numbers generated are reproducible. By setting a seed, you can get the same random numbers every time you run the code, which is useful for debugging and testing purposes.
np.random.seed(0)
#generate a random dataset 
df = np.random.exponential(scale=2, size=1000)
#convert the data into a dataframe
df = pd.DataFrame(df,columns = ['Values'])
df.head()
```

```python
#plot a histogram using seaborn 
#why i use kde=true in this code?
#KDE (Kernel Density Estimation) is a method for estimating the probability density function of a random variable. By setting kde=True, you are asking seaborn to overlay a kernel density estimate on top of the histogram. This provides a smoothed version of the histogram, which can help visualize the distribution of the data more clearly.
sns.histplot(df['Values'], kde=True)
```

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import QuantileTransformer
 
pt_boxcox = PowerTransformer(method='box-cox')

pt_yeoj = PowerTransformer(method='yeo-johnson')

qt_normal = QuantileTransformer(output_distribution='normal')

#fit the scaler to the data
df['Box_Cox'] = pt_boxcox.fit_transform(df[['Values']])

df['Yeo_Johnson']=pt_yeoj.fit_transform(df[['Values']])

df['Quantile_Tranformation']=qt_normal.fit_transform(df[['Values']])
df.head()
```

```python
# Create histogram plots for Box-Cox and Yeo-Johnson transformed data
sns.histplot(df['Box_Cox'], kde=True)

plt.title('Box-Cox Transformed Data')
plt.show()



```

```python
sns.histplot(df['Yeo_Johnson'], kde=True)
plt.title('Yeo-Johnson Transformed Data')
plt.show()
```

```python

sns.histplot(df['Quantile_Tranformation'], kde=True)
plt.title('Quantile Normal Transformed Data')
plt.show()
```

```python
#create histogram for all columns using sns.histplot use a for loop
for col in df.columns:
    sns.histplot(df[col], kde=True)
    plt.show()
```

Normalization  L1 AND L2

```python
#what is Normalization?
#Normalization is the process of scaling individual samples to have unit norm. It is often used when you want to ensure that each sample contributes equally to the analysis, regardless of its magnitude. Normalization is particularly useful in text classification and clustering tasks.
# Import necessary libraries
from sklearn.preprocessing import Normalizer
# Create a sample dataset
data = [[1,1,1],[1,1,0],[1,0,0]]
# Create a Normalizer object
normalizer = Normalizer(norm='l2')  # 'l2' for L2 normalization, 'l1' for L1 normalization
#why we use norm='l2' in this code?
#L2 normalization scales the data such that the sum of the squares of the values is equal to 1. It is commonly used in machine learning algorithms that require normalized input data, such as support vector machines and neural networks. L1 normalization, on the other hand, scales the data such that the sum of the absolute values is equal to 1. It is less common but can be useful in certain situations.
# Fit the normalizer to the data and transform it
normalized_data = normalizer.fit_transform(data)
# Display the normalized data
print(normalized_data)

```
